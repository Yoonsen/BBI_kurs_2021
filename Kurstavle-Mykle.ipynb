Vi ser litt på korpus

```python
import dhlab.nbtext as nb
```

```python
korpus = nb.book_corpus(author = 'Mykle,%agn%', limit = 200, period = (1940, 1990), ddk="839%")
korpus
```

Korpuset kan lagres til disk, og redigeres i Excel eller annet passende.

```python
korpus['urn']
```

```python
text = nb.get_corpus_text(list(korpus['urn']))
text
```

Se bort ifra de her linjene

```python
kolonner = ['-'.join([str(x[0]), str(x[1])]) for x in zip(text.columns, korpus['title'])]
```

```python
text.columns = kolonner
```

```python
text = text.fillna(0)
```

Her er litt mer meningsfulle kolonnenavn

Dokument-term-matrise (DTM)

```python
text.head(10)
```

Visualisere matrisen

```python
text.head(10).style.background_gradient()
```

Ønske å se på spesielle ord og fordelingen i korpuset

```python
torso = ['halsen', 'skuldrene', 'skulderen', 'brystene', 'brystet', 'magen', 'maven', 'midjen', 'hoftene', 'hoften','rumpen', 'lårene', 'låret']
```

```python
text.loc[torso].style.background_gradient()
```

```python
snitt = text.mean(axis=1)
```

```python
sums = text.sum(axis = 1)
```

```python
nb.frame(snitt[torso]*100/snitt[torso].sum(), 'Mykle')
```

```python
nb.frame(sums[torso]*100/sums[torso].sum(), 'Mykle')
```
