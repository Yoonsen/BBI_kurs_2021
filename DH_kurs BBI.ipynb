```python
import dhlab.nbtext as nb
```

Lag korpuset med en kommando

```python
knerten = nb.book_corpus(words = ["Knerten"], period = (1960, 1990), gender = 'f', ddk = "839%")
knerten
```

Lagre korpuset for redigering

```python
knerten.to_csv('knerten.csv')
```

```python
text = nb.get_corpus_text(list(knerten.urn))
```

Ikke nødvendig å lese...

```python
kolonner = ['-'.join([str(x[0]), str(x[1])]) for x in zip(text.columns, knerten['title'])]
```

```python
text = text.fillna(0)
```
