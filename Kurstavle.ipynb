## DH-kurs

Vi ser litt på korpus

```python
import dhlab.nbtext as nb
```

```python
knerten = nb.book_corpus(words = ['Knerten'], period = (1960, 1990), ddk="839%")
knerten
```

Korpuset kan lagres til disk, og redigeres i Excel eller annet passende.

```python
knerten.to_excel('knertenkorpus.xls')
```

```python
knerten['urn']
```

```python
text = nb.get_corpus_text(list(knerten['urn']))
text
```

Se bort ifra de her linjene

```python
kolonner = ['-'.join([str(x[0]), str(x[1])]) for x in zip(text.columns, knerten['title'])]
```

```python
text.columns = kolonner
```

Her er litt mer meningsfulle kolonnenavn

Dokument-term-matrise (DTM)

```python
text.head(10)
```

Visualisere matrisen

```python
text.head(10).style.background_gradient()
```

Ønske å se på spesielle ord og fordelingen i korpuset

```python
text.loc[['vitne', 'barn']].fillna(0).style.background_gradient()
```

```python
text.loc[['han', 'hun','jeg']].fillna(0).style.background_gradient()
```

```python
text.loc[['kirken','Kirken', 'kirke','Jesus']].fillna(0).style.background_gradient()
```

```python
text.loc[['slag','spark', 'gråt','slåss', 'kamp','slåsskamp']].fillna(0).style.background_gradient()
```

```python
summer = text.sum(axis = 1)
```

```python
summer.loc[['slag','spark', 'gråt','slåss','slåssing', 'kamp','slåsskamp']]
```

Vi kan se på konteksten til ordene i korpuset. For å sjekke betydning.

```python
nb.urn_concordance(word ='energi', urns = list(knerten['urn']))
```

```python
nb.urn_concordance(word ='slåsskamp', urns = list(knerten['urn']), before = 12, after = 12)
```

Vi kan finne relaterte ord med en passende kommando

```python
nb.relaterte_ord('vold')
```
