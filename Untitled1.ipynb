```python
import dhlab.nbtext as nb
```

```python
knerten = nb.book_corpus(words = ["Knerten"], period = (1960, 1990), gender = 'f', ddk = "839%")
knerten
```

```python
knerten_coll = nb.urn_coll(word = 'Knerten', urns = list(knerten['urn']), before = 0, after = 3)

lillebror_coll = nb.urn_coll(word = 'Lillebror', urns = list(knerten['urn']), before = 0, after = 3)

aurora_coll = nb.urn_coll(word = 'Aurora', urns = list(knerten['urn']), before = 0, after = 3)

guro_coll = nb.urn_coll(word = 'Guro', urns = list(knerten['urn']), before = 0, after = 3)
```

```python
knerten_coll.head(30)
```

```python
lillebror_coll.head(20)
```

```python
#Slår dem sammen...

import pandas as pd

alle = pd.concat([knerten_coll, lillebror_coll, aurora_coll, guro_coll], axis = 1, sort = True).fillna(0)

alle.columns = ['Knerten', 'Lillebror', 'Aurora', 'Guro']

alle.head()
```

```python
knerten_a = nb.frame_sort(nb.frame(knerten_coll[0]**1.01/alle.sum(axis = 1)).dropna())

lillebror_a = nb.frame_sort(nb.frame(lillebror_coll[0]**1.01/alle.sum(axis = 1)).dropna())

aurora_a = nb.frame_sort(nb.frame(aurora_coll[0]**1.01/alle.sum(axis = 1)).dropna())

guro_a = nb.frame_sort(nb.frame(guro_coll[0]**1.01/alle.sum(axis = 1)).dropna())
```
